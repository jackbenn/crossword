```python
from bs4 import BeautifulSoup
import requests
from pymongo import MongoClient
import time
```

```python
base_url = "https://nytcrosswordanswers.org/nyt-crossword-answers-"
```

Getting and inserting the raw pages

```python
client = MongoClient('localhost', 27017)
db = client.crossword
pages = db.pages

for date in ['02-17-21', '02-16-21', '02-15-21']:
    r = requests.get(base_url + date)
    print(r.status_code)
    pages.insert_one({date: r.content})
```
